```python
import math
def col_mx_mn(data):
  nof = np.shape(data)[1] - 1;
  val = np.zeros((nof, 4));
  in1 = np.where(data[:, -1] == 0);
  in2 = np.where(data[:, -1] == 1);

  for i in range(0, nof):
    fv = data[:, i]
    fv0 = fv[in1[0]]
    fv1 = fv[in2[0]]
    val[i,0] = np.mean(fv0) - np.std(fv0)*1.96/math.sqrt(len(in1[0]))
    val[i,1] = np.mean(fv0) + np.std(fv0)*1.96/math.sqrt(len(in1[0]))
    val[i,2] = np.mean(fv1) - np.std(fv1)*1.96/math.sqrt(len(in2[0]))
    val[i,3] = np.mean(fv1) + np.std(fv1)*1.96/math.sqrt(len(in2[0]))
  return val;
```

```python
#calculating the ci values for individual feature
def civfeat(fv, jobd):
  in1 = np.where(jobd == 0)
  in2 = np.where(jobd == 1)

  fv0 = fv[in1[0]]
  fv1 = fv[in2[0]]

  val = np.zeros((2,2))
  val[0,0] = np.mean(fv0) - np.std(fv0)*1.96/math.sqrt(len(in1[0]))
  val[1,0] = np.mean(fv0) + np.std(fv0)*1.96/math.sqrt(len(in1[0]))
  val[0,1] = np.mean(fv1) - np.std(fv1)*1.96/math.sqrt(len(in2[0]))
  val[1,1] = np.mean(fv1) + np.std(fv1)*1.96/math.sqrt(len(in2[0]))
  return val;
```

```python
def zvalue(fv, jobd): #fv - feature vector
  in1 = np.where(jobd==0)
  in2 = np.where(jobd==1)
  fv0 = fv[in1[0]]
  fv1 = fv[in2[0]]
  m0 = np.mean(fv0)
  m1 = np.mean(fv1)

  if(m0==0 and m1==0):
    return 0

  var0 = np.std(fv0)
  var1 = np.std(fv1)

  zv = (m0-m1)/math.sqrt(var0*var0/len(in1[0])+var1*var1/len(in2[0]))
  return np.abs(zv)
```

```python
from statsmodels.stats.weightstats import ztest
#calculating zvalue of individual feature using ztest - predefined function
def zpython(fv, jobd):
  in1=np.where(jobd==0)
  in2=np.where(jobd==1)
  fv0=fv[in1[0]]
  fv1=fv[in2[0]]
  m0=np.mean(fv0)
  m1=np.mean(fv1)
  if m0==0 and m1==0:
    return 0,0
  zv,pv=ztest(fv0,fv1,alternative='two-sided')
  return np.abs(zv),pv
```

```python
#calculating z-cvalue for each feature using the zvalue function
from scipy import stats
def zall(data):
  zv=np.zeros((np.shape(data)[1]-1))
  pv=np.zeros((np.shape(data)[1]-1))
  for i in range(0,np.shape(data)[1]-1):
    fv=data[:,i]
    zv[i]=zvalue(fv,data[:,-1])
    pv[i]=stats.norm.sf(zv[i])*2
  return zv,pv
```

```python
#ranksum test
def rstf(fv, jobd):
  in1=np.where(jobd==0)
  in2=np.where(jobd==1)
  fv0=fv[in1[0]]
  fv1=fv[in2[0]]
  rs,pv=stats.ranksums(fv0,fv1)
  return pv
```

```python
#ranksum test for individual feature
from scipy import stats
def rst(data):
  pv=np.zeros((np.shape(data)[1]-1))
  for i in range(0,np.shape(data)[1]-1):
    fv=data[:,i]
    pv[i]=rstf(fv,data[:,-1])
  return pv
```

```python
#calculating z-value for each feature using the zpython function
def zall1(data):
  zv=np.zeros((np.shape(data)[1]-1))
  pv=np.zeros((np.shape(data)[1]-1))
  for i in range(0,np.shape(data)[1]-1):
    fv=data[:,i]
    zv[i],pv[i]=zpython(fv,data[:,-1])
  return zv,pv
```

```python
def gini(f, jobd):
  mf = np.median(f)
  in1 = np.where(f>=mf)
  in2 = np.where(f<mf)
  if len(in1[0])==0 or len(in2[0])==0:
    return 0.5

  job1 = jobd[in1[0]]
  job0 = jobd[in2[0]]

  n01 = np.sum(job0)
  n00 = len(job0)-n01

  n11 = np.sum(job1)
  n10 = len(job1)-n11

  p00 = n00 / (n00+n01)
  p01 = n01 / (n00+n01)
  p10 = n10 / (n00+n11)
  p11 = n11 / (n00+n11)

  gin0 = 1- (p00*p00) - (p01*p01)
  gin1 = 1- (p10*p10) - (p11*p11)

  fg = (len(job1)/len(f))*gin1 + (len(job0)/len(f))*gin0
  return fg
```

```python
def infogain(f, jobd):
  mf = np.median(f)
  in1 = np.where(f>=mf)
  in2 = np.where(f<mf)
  if len(in1[0])==0 or len(in2[0])==0:
    return 0.5

  job0 = jobd[in1[0]]
  job1 = jobd[in2[0]]

  n01 = np.sum(job0)
  n00 = len(in1[0])-n01

  n11 = np.sum(job1)
  n10 = len(in2[0])-n11

  p00 = n00 / (n00+n01)
  p01 = n01 / (n00+n01)
  p10 = n10 / (n11+n10)
  p11 = n11 / (n11+n10)

  n0 = np.sum(jobd)
  n1 = len(jobd) - n0
  p0 = n0 / len(jobd) if (n0+n1) != 0 else 0
  p1 = n1 / len(jobd) if (n0+n1)!= 0 else 0

  en0 = -p00*np.log2(p00) - p01*np.log2(p01)
  en1 = -p10*np.log2(p10) - p11*np.log2(p11)

  # fg = (len(job1)/len(f))*gin1 + (len(job0)/len(f))*gin0
  # return fg


  parent_en = -p0*np.log2(p0) - p1*np.log2(p1)
  childen0 = -p00*np.log2(p00)- p01*np.log2(p01)
  childen1 = -p11*np.log2(p11) - p10*np.log2(p10)

  info_gain = parent_en - (len(in1[0])/len(f))*childen0 - (len(in2[0])/len(f))*childen1
  # print(info_gain)
  return info_gain
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ranksums
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from matplotlib import pyplot

def normalizedata(X_train):
    scaler = MinMaxScaler()
    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    return X_train
```

```python
def crossover(chr):
  no_of_chr = np.shape(chr)[0]
  np.random.shuffle(chr)
  new_chr = np.zeros((int(no_of_chr/2), np.shape(chr)[1]))
  for i in range(0, int(no_of_chr/2)):
    new_chr[i,0:10] = chr[2*i, 0:10] #taking the first 10 features of the 2ith chr
    new_chr[i,10:] = chr[2*i+1, 10:] #next 10 features of the 2i+1 th chr  -- this creates a new chromosome based on two parents
  return new_chr
```

```python
def mutation(chr):
  change = np.random.randint(0, np.shape(chr)[1], [np.shape(chr)[0], 2])
  for i in range(0, np.shape(chr)[0]):
    chr[i, change[i, 0]] = 1-chr[i, change[i, 0]] #making all zeros to ones and ones to zeros
    chr[i, change[i, 1]] = 1-chr[i, change[i, 1]]
  return chr
```

```python
#genetic algo
def GA(data):
  chr = np.random.randint(0,2,[50, np.shape(data)[1]-1])
  # print('accuracy with all features = ' + str(accuracy_value(data)))
  best_fit = 0 #best fitness value
  gen = 0 #iteration or generation
  while(gen <= 20):
    fit = fitness(chr, data)
    a = np.argmax(fit[:, 2])

    current_best_fit = fit[a, 2]
    if current_best_fit > best_fit:
      best_fit = current_best_fit
      best_accuracy = fit[a, 0]
      best_nof_features = fit[a, 1]
      best_chr = copy(chr[a, :])

    chr[0:25, :] = crossover(chr) #replacing our original 50 chromosomes to 25 with crossover
    chr[0:25, :] = mutation(chr[0:25, :]) #appliies mutation on above generated cross-over chrs and replace them from 0-25 again
    chr[25:, :] =  np.random.randint(0,2,[25, np.shape(chr)[1]])   #regeneration

    in1 = np.where(best_chr==1)
    best_fts = data[:, in1[0]]

    gen = gen + 1
    # print('best acc = ',best_accuracy,'bnof = ',best_nof_features, "best_chrs", best_chr)
  return best_fts
```
